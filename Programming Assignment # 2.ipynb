```python
import numpy as np

X = np.random.rand(5,5)

mean_X = X.mean()
std_X = X.std()

X_normalized = (X - mean_X) / std_X

np.save('X_normalized.npy', X_normalized)

X, X_normalized
```

```python
import numpy as np

A = np.arange(1, 101).reshape(10, 10) ** 2

div_by_3 = A[A % 3 == 0]

np.save('div_by_3.npy', div_by_3)

A, div_by_3
```
